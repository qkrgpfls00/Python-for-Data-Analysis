```python
import numpy as np
import random
```

```python
#백 번 해보기

position=0 #현위치
walk=[position] #계단을 오르거나(+1) 내려간(-1) 값의 누적합
steps=1000
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position +=step
    walk.append(position)
```

```python
import matplotlib.pyplot as plt
plt.plot(walk[:100])
```

```python
#천 번 해보기

nsteps=1000
draws=np.random.randint(0,2,size=nsteps)
steps=np.where(draws>0,1,-1)
walks=steps.cumsum()
```

```python
plt.plot(walks)
```

```python
#한 번에 해보기
nwalks=5000
nsteps=1000
draws=np.random.randint(0,2,size=(nwalks, nsteps)) #0또는 1
steps=np.where(draws>0,1,-1)
walks=steps.cumsum(1)
walks
```

```python
walks.max() #최댓값
```

```python
walks.min() #최솟값
```

```python
#누적합이 30또는 -30이 되는 최소시점 찾아보기

hits30 =(np.abs(walks)>=30).any(1) 
hits30
```

```python
hits30.sum() #누적합이 30또는 -30이 되는 경우
```

```python
crossing_times=(np.abs(walks[hits30])>=30).argmax(1) #walks에서 컬럼 선택 후 절댓값 30이 넘는 경우에 대해 축 1의 argmax값 구하기
crossing_times
```

```python
crossing_times.mean()
```
