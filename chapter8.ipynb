# 데이터 합치기

```python
# pandas.merge는 하나 이상의 키를 기준으로 DF의 로우를 합친다. SQL이나 다른 관계형 데이터베이스의 join과 유사
# pandas.concat은 하나의 축을 따라 객체를 이어붙임
# combile_first 인스턴스 메서드는 두 객체를 포개어 한 객체에서 누락된 데이터를 다른 객체에 있는 값으로 채울 수 있도록 함
```

* 데이터베이스 스타일로 DataFrame 합치기

```python
# 다대일의 경우

df1=pd.DataFrame({'key':['b','b','a','c','a','a','b'], 
                 'data1':range(7)})
df1 # key컬럼에 여러 개의 a,b 로우를 가지고 있음
```

```python
df2=pd.DataFrame({'key':['a','b','d'],
                 'data2':range(3)})
df2 # key컬럼에 유일한 로우를 가지고 있음
```

```python
pd.merge(df1,df2, on='key') # merge함수는 중복된 컬럼 이름을 키로 사용하지만, 명시적으로 지정해주는 습관을 들이자 (on='key')
```

```python
# 두 객체에 중복된 컬럼 이름이 하나도 없으면 따로 지정

df3=pd.DataFrame({'lkey':['b','b','a','c','a','a','b'],
                 'data1':range(7)}) 
df3
```

```python
df4=pd.DataFrame({'rkey':['a','b','d'],
                 'data2':range(3)})
df4
```

```python
pd.merge(df3,df4,left_on='lkey', right_on='rkey') # merge함수는 내부조인을 수행하여 교집합인 결과 반환
```

```python
pd.merge(df1,df2, how='outer') # how인자로 left(왼쪽조인) ,right(오른쪽조인) ,outer(외부조인) 수행 가능
```

```python
# 다대다 병합

df1=pd.DataFrame({'key':['b','b','a','c','a','b'],
                 'data1':range(6)})
df1
```

```python
df2=pd.DataFrame({'key':['a','b','a','b','d'],
                 'data2':range(5)})
df2
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,how='inner') # 다대다 조인은 두 로우의 데카르트곱을 반환
```

```python
left=pd.DataFrame({'key1':['foo','foo','bar'],
                  'key2':['one','two','one'],
                  'lval':[1,2,3]})
right=pd.DataFrame({'key1':['foo','foo','bar','bar'],
                   'key2':['one','one','one','two'],
                   'rval':[4,5,6,7]})
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer') # 여러 개의 키를 병합하려면 컬럼 이름이 담긴 리스트를 넘기기
```

```python
# 컬럼 이름에 대한 처리

pd.merge(left,right,on='key1')
```

```python
pd.merge(left,right,on='key1',suffixes=('_left','_right')) # suffixes 인자로 두 객체에서 겹치는 컬럼 이름 뒤에 붙일 문자열 지정
```

* 색인 병합하기

```python
left1=pd.DataFrame({'key':['a','b','a','a','b','c'],
                   'value':range(6)})
left1
```

```python
right1=pd.DataFrame({'group_val':[3.5,7]},
                   index=['a','b'])
right1
```

```python
pd.merge(left1, right1, left_on='key',right_index=True)
```

```python
pd.merge(left1,right1,left_on='key',right_index=True,how='outer') # 병합은 기본적으로 교집합, 외부조인을 실행하여 합집합
```

```python
# 계층 색인된 데이터는 암묵적으로 여러 키를 병합하는 것

lefth=pd.DataFrame({'key1':['Ohio','Ohio','Ohio','Nevada','Nevada'],
                   'key2':[2000,2001,2002,2001,2002],
                   'data':np.arange(5.)})
lefth
```

```python
righth=pd.DataFrame(np.arange(12).reshape((6,2)),
                   index=[['Nevada','Nevada','Ohio','Ohio','Ohio','Ohio'],
                         [2001,2000,2000,2000,2001,2002]],
                   columns=['event1','event2'])
righth
```

```python
pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True) # 리스트로 여러 개의 컬럼을 지정해서 병합
```

```python
pd.merge(lefth,righth,left_on=['key1','key2'],
        right_index=True, how='outer') # 중복되는 색인값을 다룰 때는 how='outer'옵션을 사용해야한다.
```

```python
# 양쪽에 공통적으로 존재하는 여러 개의 색인을 병합하는 것도 가능
left2=pd.DataFrame([[1.,2.],[3.,4.],[5.,6.]],
                  index=['a','c','e'],
                  columns=['Ohio','Nevada'])
left2
```

```python
right2=pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[13,14]],
               index=['b','c','d','e'],
               columns=['Missouri','Alabama'])
right2
```

```python
pd.merge(left2,right2,how='outer',left_index=True,right_index=True)
```

```python
# 색인으로 병합할 때 DataFrame의 join 메서드를 사용하면 편리

left2.join(right2, how='outer') # 컬럼이 겹치지 않으며 완전히 같거나 유사한 색인 구조를 가진 객체들을 병합할 떄
```

```python
left1.join(right1,on='key')
```

```python
# 색인 대 색인으로 두 DF을 병합하려면 DF의 리스트를 join메서드로 넘기면 됨, concat사용

another=pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[16.,17.]],
                    index=['a','c','e','f'],
                    columns=['New York','Oregon'])
another
```

```python
left2.join([right2,another])
```

```python
left2.join([right2,another],how='outer')
```

# 재형성과 피벗

* 계층적 색인으로 재형성하기

```python
data=pd.DataFrame(np.arange(6).reshape((2,3)),
                 index=pd.Index(['Ohio','Colorado'],name='state'),
                 columns=pd.Index(['one','two','three'],
                                 name='number'))
data
```

```python
result=data.stack() # stack 메서드를 사용하면 컬럼이 로우로 피벗되어 Series객체 반환
result
```

```python
result.unstack() # unstack 메서드로 계층적 색인을 가진 Series로부터 다시 DataFrame을 얻음
```

```python
result.unstack(0)
```

```python
result.unstack('state')
```

```python
# 해당 레벨에 있는 모든 값이 하위그룹에 속하지 않을 경우 unstack을 하게되면 누락된 데이터 생성 가능

s1=pd.Series([0,1,2,3],index=['a','b','c','d'])
s2=pd.Series([4,5,6], index=['c','d','e'])
```

```python
data2=pd.concat([s1,s2],keys=['one','two'])
data2
```

```python
data2.unstack()
```

```python
data2.unstack().stack() # stack 메서드는 누락된 데이터를 자동으로 걸러냄
```

```python
data2.unstack().stack(dropna=False)
```

```python
df=pd.DataFrame({'left':result, 'right':result+5},
               columns=pd.Index(['left','right'], name='side'))
df
```

```python
df.unstack('state')
```

```python
df.unstack('state').stack('side') # stack을 호출할 때 쌓을 축의 이름을 지정랑 수 있음
```

* 긴 형식에서 넓은 형식으로 피벗하기: pivot

```python
data=pd.read_csv('examples/macrodata.csv')
```

```python
data.head()
```

```python
periods=pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
```

```python
columns=pd.Index(['realgdp','infl','unemp'],name='item')
```

```python
data=data.reindex(columns=columns)
```

```python
data.index=periods.to_timestamp('D','end')
```

```python
ldata=data.stack().reset_index().rename(columns={0:'value'})
```

```python
ldata[:10] # 긴 형식: 여러 시계열이나 둘 이상의 키를 가지고 있는 다른 관측 데이터에서 사용, 각 로우는 단일 관측치를 나타냄
```

```python
pivoted=ldata.pivot('date','item','value') 
pivoted # 긴 형식은 작업에 용이하지 않아 pivot메서드로 하나의 DF에 date컬럼의 시간값으로 색인된 개별 item을 컬럼으로 포함
```

```python
ldata['value2']=np.random.randn(len(ldata))
ldata[:10]
```

```python
# pivot메서드의 처음 두 인자는 로우와 컬럼 색인으로 사용될 컬럼이름, 마지막 두 인자는 DF에 채워 넣을 값을 담고 있는 컬럼 이름
pivoted=ldata.pivot('date','item')
pivoted[:5]
```

```python
pivoted['value'][:5]
```

```python
# pivot은 set_index를 사용하여 계층적 색인을 만들고 unstack메서드를 사용하여 형태를 변형하는 단축키 같은 역할
unstacked=ldata.set_index(['date','item']).unstack('item')
unstacked[:7]
```

* 넓은 형식에서 긴 형식으로 피벗하기: melt

```python
df=pd.DataFrame({'key':['foo','bar','baz'],
                'A':[1,2,3],
                'B':[4,5,6],
                'C':[7,8,9]})
df
```

```python
melted=pd.melt(df,['key']) # 'key'컬럼을 그룹 구분자로 사용하거나 다른 컬럼을 데이터 값으로 사용할 수 있음
melted # 그룹 구분자로 사용
```

```python
reshaped=melted.pivot('key','variable','value') # pivot을 사용해 원래 모양으로 되돌리기-> 로우 라벨로 사용하던 컬럼에서 색인 생성
reshaped
```

```python
reshaped.reset_index() # reset_index를 사용해 다시 컬럼으로 돌려놓기
```

```python
pd.melt(df, id_vars=['key'], value_vars=['A','B']) # 데이터값으로 사용할 컬럼들의 집합 지정
```

```python
pd.melt(df, value_vars=['A','B','C'])
```

```python
pd.melt(df, value_vars=['key','B','C']) # 그룹 구분자 없이도 사용
```
